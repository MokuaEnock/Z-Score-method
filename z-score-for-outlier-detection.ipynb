## Z-Score method.
* The z-score is also reffered to as the **standard score**.
*  It helps us conceptualize how many standard deviations away a data value is from the mean and how bi or small the value is than the mean.
*  Mathematically the z-score is given by the formula **Z = (x - mean)/y**

> Where x = data value and y = standard deviation 

#### Z-Score and outliers.
* During the computation of teh z-score we expect the z-score of the data to fall within the range of **-3 < x > 3** if it falls out of this range this indicates that the data point is quite unique and this value might be an outlier.

### Examples.

Lets generate random data and walk the talk.

```python
#importing libraries for use.
import numpy as np
import random

#generating random data for the students 
data = [random.randint(0, 100) for i in range(10)]
print(data)
```

```python
#calculating the mean and standard deviation of the data
dmean  = np.mean(data)
dstd = np.std(data)
print ('Mean of the data is : ' + str(dmean))
print ('Standard deviation of the data is : ' + str(dstd))
```

```python
#calculating the z zscore with the threshold at 3
threshold = 3
outlier = []
for i in data:
    z = (i - dmean) / dstd
    if z > threshold:
        outlier.append(i)
print ('outlier in the dataset is', outlier)    
```

* Another way of calculating the z-score is using the scipy module

```python
#using scipy.stats
from scipy import stats
stats.zscore(data, axis = None)
```

```python
# Using a Sample Pandas Dataframe
import pandas as pd

df = pd.DataFrame.from_dict({
    'Name': ['Nik', 'Kate', 'Joe', 'Mitch', 'Alana'],
    'Age': [32, 30, 67, 34, 20],
    'Income': [800000, 9000, 4500, 2300, 12000],
    'Education' : [5, 7, 3, 4, 4]
})
df.head()
```

```python
#creating a column with the normalized values from the income column 
df['Income Zscore'] = stats.zscore(df.Income)
df.head()
```

```python
#converting our dataframe columns into zscores using pandas.apply() method
df = df.select_dtypes(include = 'number').apply(stats.zscore).drop(columns = ['Income Zscore'])
df.head()
```
